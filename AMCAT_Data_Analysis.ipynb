### Lets remove outliers from all the columns which have outliers

```python
# Following IQR based method to remove outliers.
def get_IQRValues(df,column):
    q1, q3 = df[column].quantile(q = [0.25, 0.75]) 
    iqr = q3-q1 
    ub = q3+1.5*iqr
    lb = q1-1.5*iqr
    return {"q1" : q1, "q3":q3, "iqr": iqr, "UB": ub, "LB": lb}
```

### Check for outliers in `CollegeTier` column

```python
ct_dict = get_IQRValues(amcat_df_Sal,"CollegeTier")
ct_dict["UB"]
```

```python
amcat_df_Sal[amcat_df_Sal.CollegeTier < ct_dict["UB"]].CollegeTier.value_counts()
```
